### 1. Number of URLH which are overlapping.

NOTE - I have assumed that you meant 'urlh' and not 'URLH'.
Assumption - Overlapping means intersection(analogy with intersection of 2 sets).

```python
# Assumption - All the unique urlhs in both today and yesterday's data.

inter=list(set(a['urlh']).intersection(set(b['urlh'])))
inter[0:5]
```

```python
len(inter)
```

```python
# NOTE- There are different number of 'urlh' in both file which are overlapping.
# REASON - There is multiplicity in the number of 'urlh' values.

overlap_today=[]

for i in tqdm(range(len(a))):
    if(a['urlh'][i] in inter):
        overlap_today.append(a['urlh'][i])
        
        
```

```python
overlap_today[0:5]
```

```python
len(overlap_today)
```

```python
overlap_yesterday=[]

for i in tqdm(range(len(b))):
    if(b['urlh'][i] in inter):
        overlap_yesterday.append(b['urlh'][i])
```

```python
len(overlap_yesterday)
```

### 2. Calculate price difference.

```python
a['urlh'][1]
```

```python
my_dict_t={}

for i in tqdm(range (len(a))):
    if(a['urlh'][i] in overlap_today):
        if(a['urlh'][i] in my_dict_t):
            if(a['http_status'][my_dict_t[a['urlh'][i]]]==200):
                continue
            else:
                my_dict_t.pop(a['urlh'][i])
                my_dict_t[a['urlh'][i]]=i
        else:
            my_dict_t[a['urlh'][i]]=i
        
```

```python
len(my_dict_t)
```

```python
b['urlh'][1]
```

```python
my_dict_y={}

for i in tqdm(range (len(b))):
    if(b['urlh'][i] in overlap_yesterday):
        if(b['urlh'][i] in my_dict_y):
            if(b['http_status'][my_dict_y[b['urlh'][i]]]==200):
                continue
            else:
                my_dict_y.pop(b['urlh'][i])
                my_dict_y[b['urlh'][i]]=i
        else:
            my_dict_y[b['urlh'][i]]=i
            
        
```

```python
len(my_dict_y)
```

```python
# NOTE - As the price difference can only be calculated for not None values thus removing those fields.
        
price_diff=[]

for i in tqdm(my_dict_t):
    if((a['available_price'][my_dict_t[i]] is not None) and (b['available_price'][my_dict_y[i]] is not None)):
        price_diff.append(float(a['available_price'][my_dict_t[i]])-float(b['available_price'][my_dict_y[i]]))
```

### 5(b). Generate a new file where mrp is normalized. If there is a 0 or a non-float value or the key doesn't exist, make it Nan.

Assumption - A new file where 'mrp' is normalized means you are talking of today.json file. Also the same thing that I did with today file can be done on yesterday.json.  

```python
mrp_col = a['mrp']
mrp_list = [float(x) for x in mrp_col if x != None]
x_max = max(mrp_list)
x_min = min(mrp_list)
```

```python
def normalize(x):
    if x != None:
        x = float(x)
    else:
        return 'NA'
    s = x_max -x_min
    x_new = (x - x_min) / s
    return x_new

Normalized = a
Normalized['mrp'] = Normalized['mrp'].apply(lambda x : normalize(x))
```

```python
Normalized.dtypes
```

```python
Normalized.mrp = Normalized.mrp.astype(str)
```

```python
Normalized.dtypes
```

```python
Normalized['mrp']=Normalized['mrp'].replace({'0.0':"NA",0:"NA"})
```

```python
with open('Normalized.json', 'w') as f:
    f.write(Normalized.to_json(orient='records', lines=True))
```
